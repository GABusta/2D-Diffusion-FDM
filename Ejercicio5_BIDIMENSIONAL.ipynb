$$

$$
\frac{\partial^2 T}{\partial y^2}=\frac{T_{i,j+1}-2.T_{i,j}+T_{i,j-1}}{\Delta h_{y}^2}+ o(h_{y}^2)
$$

y Discretizando el tiempo,
$$
\frac{\partial T}{\partial t}=\frac{T_{i,j}^{n+1} -T_{i,j}^n}{\Delta t}+ o(\Delta t)

$$

## Programa DIFUSIÓN BIDIMENSIONAL - Transitorio

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import mpl_toolkits.mplot3d
```

```python

#------------------------------------------
#------- Constantes y Cond. Inic. ---------
#------------------------------------------ 
kx,ky=72.2,72.2  # Constante de Difusividad
ro=7870.0        # Densidad del material
cp=440.0         # Calor específico

L=1.0/100        # [m] Largo de la placa
Np=30            # Cantidad de elementos por lado
dx=L/Np          # Delta posición   dx = dy

t=150.0           # [seg.] Tiempo de simulación
Nt=400            # Intervalos de tiempo 
dt=t/Nt           # Delta de Tiempo

Q=0.0            # Generación Interna de calor 

A=kx*dt/(ro*cp*dx**2)
B=ky*dt/(ro*cp*dx**2)
C=Q*dt/(ro*cp)

Nodos=(Np+1)**2
m=np.zeros((Nodos,Nodos)) # Inicializo Matriz de coeficientes
R=np.zeros(Nodos)         # Inicializo Vector de resultados
T=np.zeros(Nodos)         # Inicializo TEMPERATURAS
T[:]=273.0                # Temp. Inicial de la PLACA

T1,T2,T3=300.0,300.0,1000.0 # Temp. Impuestas Bordes (Ver figura)

#------------------------------------------
#---------- Ensamble MATRIZ  --------------
#------------------------------------------
for i in range(1,Nt+1):          # Recorre el tiempo
    for j in range(1,Np+2):      # Recorre las FILAS
        for k in range(1,Np+2):  # Recorre las COLUMNAS
            p=(j-1)*(Np+1) + k -1
            
            #-------------------------------
            #--------- PRIMER FILA ---------
            #-------------------------------
            if j==1:
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C -A*T1 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C -A*T3 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C
            
            #-------------------------------
            #-------- ULTIMA FILA ----------
            #-------------------------------
            elif j==Np+1:
                R[p]=T2
                m[p,p]=1.0
                          
            #-------------------------------
            #----- ANTE ULTIMA FILA --------
            #-------------------------------
            elif j==Np:
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C -A*T1 -B*T2 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C -A*T3 -B*T2 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C -B*T2
            
            #-------------------------------
            #------ FILAS INTERMEDIAS ------
            #-------------------------------
            else:               
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p+Np+1]=B
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C -A*T1 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+Np+1]=B
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C -A*T3 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p+(Np+1)]=B
                    m[p,p-(Np+1)]=B
                    R[p]= -T[p] -C
                       
    Resul=np.linalg.inv(m).dot(R)  
    T=Resul 
# Para cada dt, se calculan las temperaturas y esas mismas se utilizan para calcular las Temperaturas
# Nodales un instante de tiempo después
```

```python
x = np.linspace(0.0, L, Np+1) 
y = np.linspace(0.0, L, Np+1) 
X, Y = np.meshgrid(x, y) # Grilla x-y
 
Z=T.reshape(Np+1,Np+1)   # Rediseño el vector TEMPERATURA NODAL

cs = plt.contourf(X, Y, Z) 
cbar = plt.colorbar(cs) 
  
plt.title('Temperatura Transitoria - placa plana') 
plt.show() 
```

# 2. DIFUSIÓN DOMINIO AXISIMÉTRICO

$$
\frac{\partial}{\partial z}.(k_{z}.\frac{\partial T}{\partial z})+\frac{1}{r}.\frac{\partial}{\partial r}.(r.k_{r}.\frac{\partial T}{\partial r}) + Q = \rho.C.\frac{\partial T}{\partial t}
$$
$$
$$
$$
k_{z}.\frac{\partial ^2 T}{\partial z^2} + \frac{k_{r}}{r}.\frac{\partial T}{\partial r} + k_{r}.\frac{\partial ^2 T}{\partial r^2} + Q = \rho.C.\frac{\partial T}{\partial t}
$$
$$
$$

$$

$$
$$
$$
$$
$$
\frac{\partial T}{\partial r}=\frac{T_{i,j+1}-T_{i,j-1}}{\Delta h_{r}}+ o(h_{r})
$$

$$

$$

## Programa DIFUSIÓN AXISIMÉTRICO - Transitorio

```python
#------------------------------------------
#------- Constantes y Cond. Inic. ---------
#------------------------------------------ 
kz,kr=72.2,72.2  # Constante de Difusividad
ro=7870.0        # Densidad del material
cp=440.0         # Calor específico

L=1.0/100        # [m] Largo de la placa
Np=30            # Cantidad de elementos por lado
dz=L/Np          # Delta posición   dz = dr

t=150.0           # [seg.] Tiempo de simulación
Nt=400            # Intervalos de tiempo 
dt=t/Nt           # Delta de Tiempo

Q=0.0            # Generación Interna de calor 

A=kz*dt/(ro*cp*dz**2)
B=kr*dt/(ro*cp*dz**2)
C=Q*dt/(ro*cp)
D=kr*dt/(L*ro*cp*dz*2)

Nodos=(Np+1)**2
m=np.zeros((Nodos,Nodos)) # Inicializo Matriz de coeficientes
R=np.zeros(Nodos)         # Inicializo Vector de resultados
T=np.zeros(Nodos)         # Inicializo TEMPERATURAS
T[:]=273.0                # Temp. Inicial de la PLACA

T1,T2,T3=300.0,300.0,1000.0 # Temp. Impuestas Bordes (Ver figura)

#------------------------------------------
#---------- Ensamble MATRIZ  --------------
#------------------------------------------
for i in range(1,Nt+1):          # Recorre el tiempo
    for j in range(1,Np+2):      # Recorre las FILAS
        for k in range(1,Np+2):  # Recorre las COLUMNAS
            p=(j-1)*(Np+1) + k -1
            
            #-------------------------------
            #--------- PRIMER FILA ---------
            #-------------------------------
            if j==1:
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C -A*T1 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C -A*T3 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p+Np+1]=2*B
                    R[p]= -T[p] -C
            
            #-------------------------------
            #-------- ULTIMA FILA ----------
            #-------------------------------
            elif j==Np+1:
                R[p]=T2
                m[p,p]=1.0
                          
            #-------------------------------
            #----- ANTE ULTIMA FILA --------
            #-------------------------------
            elif j==Np:
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C -A*T1 -(B+D)*T2 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C -A*T3 -(B+D)*T2 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C -(B+D)*T2
            
            #-------------------------------
            #------ FILAS INTERMEDIAS ------
            #-------------------------------
            else:               
                if k==1:
               #---primer columna---
                    R[p]=T1
                    m[p,p]=1.0
                elif k==Np+1:
               #---ultima columna---
                    R[p]=T3
                    m[p,p]=1.0
                elif k==2:
               #---segunda columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p+1]=A
                    m[p,p+Np+1]=B+D
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C -A*T1 
                elif k==Np:
               #---Ante ultima columna--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+Np+1]=B+D
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C -A*T3 
                else: 
               #---Columnas intermedias--- 
                    m[p,p]=-2*A -2*B -1
                    m[p,p-1]=A
                    m[p,p+1]=A
                    m[p,p+(Np+1)]=B+D
                    m[p,p-(Np+1)]=B-D
                    R[p]= -T[p] -C
                       
    Resul=np.linalg.inv(m).dot(R)  
    T=Resul 
# Para cada dt, se calculan las temperaturas y esas mismas se utilizan para calcular las Temperaturas
# Nodales un instante de tiempo después
```

```python
x = np.linspace(0.0, L, Np+1) 
y = np.linspace(0.0, L, Np+1) 
X, Y = np.meshgrid(x, y) # Grilla x-y
 
Z=T.reshape(Np+1,Np+1)   # Rediseño el vector TEMPERATURA NODAL

cs = plt.contourf(X, Y, Z) 
cbar = plt.colorbar(cs) 
  
plt.title('Temperatura Transitoria - placa Axisimétrica') 
plt.show() 
```
